# Data
The data we will be using is hourly power consumption data from PJM. Energy consumtion has some unique charachteristics. It will be interesting to see how prophet picks them up.

Pulling the `PJM East` which has data from 2002-2018 for the entire east region.

3. Plot the time series

````python
fig, ax = plt.subplots(1,1,figsize=(14,6))
sns.lineplot(data=df, x=df.index, y="PJME_MW")
plt.title("First Plot")
plt.show()
````

# EDA

4. Create this collection of new variables :
* ```date``` equal to the index
* ```hour``` equal to the hour of the day
* ```dayofweek```
* ```quarter```
* ```month```
* ```year```
* ```dayofyear```
* ```dayofmonth```

Create an object X containing the above variables and y containing the time series

````python
df["hour"] = df.index.hour
df["dayofweek"] = df.index.dayofweek
df["quarter"] = df.index.quarter
df["month"] = df.index.month
df["year"] = df.index.year
df["dayofyear"] = df.index.dayofyear
df["dayofmonth"] = df.index.day
df["weekofyear"] = df.index.isocalendar().week
````

5. Create an object features_and_target containing both and display the first few lines

````python
features_and_target = df.copy()
features_and_target.head()
````

## Plotting the Features to see trends

6. Plot the time series against hour, dayofweek, year, weekofyear


````python
sns.pairplot(features_and_target,
             hue='hour',
             x_vars=['hour','dayofweek',
                     'year','weekofyear'],
             y_vars='PJME_MW',
             height =6
            )
plt.suptitle('Power Use MW by Hour, Day of Week, Year and Week of Year')
plt.show()
````

# Train/Test Split
7. Cut off the data after 2015 to use as our validation set. We will train on earlier data.

````python
X_trian = features_and_target.loc[df["year"] < 2015]
X_test = features_and_target.loc[df["year"] >= 2015]
````

8. plot the series with different colors for the train and test set to visualize the separation.

````python
fig, ax = plt.subplots(1,1,figsize=(16,6))
sns.lineplot(x = X_trian.index, y = X_trian["PJME_MW"], color = "r", ax=ax)
sns.lineplot(x = X_test.index, y = X_test["PJME_MW"], color = "y", ax=ax)
````

# Simple Prophet Model
9. In pjme_train rename the Datetime variable "ds" and PJME_MW "y" to fit into the Prophet framework

````python
X_trian = X_trian.reset_index().rename(columns={"Datetime": "ds", "PJME_MW": "y"})
X_test = X_test.reset_index().rename(columns={"Datetime": "ds", "PJME_MW": "y"})
display(X_trian.head())
display(X_test.head())
````

10. Create an instance of the Prophet model and fit it on the train data.

````python
from fbprophet import Prophet

m = Prophet()
m.fit(X_trian)
````

11. Make predictions with your trained default prophet model on the test set

````python
forecast = m.predict(X_test)
forecast.head()
````

13. Plot the trend, the yearly seasonnality, the weekly seasonnality, and the daily seasonnality.

````python
fig = m.plot_components(forecast, figsize=(14, 16))
````

# Compare Forecast to Actuals

14. Plot the forcasts against the actual values in the test set.

````python
f, ax = plt.subplots(1,1,figsize=(16,6))
sns.lineplot(x = X_test.ds, y = X_test["y"], color = "r", ax=ax)
fig = m.plot(forecast, ax=ax)
plt.show()
````

# Look at first month of predictions

15. Compare in a graph predictions for the first month against reality.

````python
from datetime import datetime

first_month = X_test.loc[(X_test["year"] == 2015) & (X_test["month"] == 1)]
first_month_predict = forecast.loc[forecast["ds"].isin(first_month["ds"].values), :]

f, ax = plt.subplots(1,1,figsize=(16,6))
ax.scatter(first_month.ds, first_month['y'], color='r')
fig = m.plot(first_month_predict, ax=ax)
ax.set_xbound(lower=datetime.strptime('01-01-2015', "%m-%d-%Y"),
              upper=datetime.strptime('02-01-2015', "%m-%d-%Y"))
plt.suptitle('January 2015 Forecast vs Actuals')
plt.show()
````

17. Calculate the MSE, the MAE and the MAPE (mean obsolute percentage error) for this model over the test dataset
$$MAPE = \frac{1}{n}\sum_{t=1}^{n}|\frac{Y_t - \hat{Y_t}}{Y_t}|$$

````python
from sklearn.metrics import mean_squared_error, mean_absolute_error

print("The MSE on the test set is : %f" % mean_squared_error(y_true=X_test["y"], y_pred=forecast["yhat"]))
````

````python
print("The MAE on the test set is : %f" % mean_absolute_error(y_true=X_test["y"], y_pred=forecast["yhat"]))
````

````python
def mean_absolute_percentage_error(y_true, y_pred): 
    return (((y_true - y_pred) / y_true).abs()).mean() * 100

print("The MAPE on the test set is : %f" % mean_absolute_percentage_error(y_true=X_test["y"], y_pred=forecast["yhat"]))
````

# Adding Holidays
Next we will see if adding holiday indicators will help the accuracy of the model. Prophet comes with a *Holiday Effects* parameter that can be provided to the model prior to training. 

We will use the built in pandas `USFederalHolidayCalendar` to pull the list of holidays

18. Create a test_holidays object containing the us federal holidays for the considered test period and an object train_holidays for the considered train period.

````python
from pandas.tseries.holiday import USFederalHolidayCalendar as calendar

cal = calendar()
X_train_holidays = cal.holidays(start=X_trian["ds"].min(), end=X_trian["ds"].max())
X_test_holidays = cal.holidays(start=X_test["ds"].min(), end=X_test["ds"].max())
````

19. Create a DataFrame contaning all holidays over the observed period with a ds column with the dates and a holiday column with the holidays and convert ds column to datetime format.

````python
X_trian["USFederalHoliday"] = X_trian["ds"].isin(X_train_holidays)
X_test["USFederalHoliday"] = X_test["ds"].isin(X_test_holidays)
holiday_dates = pd.concat([X_trian[X_trian['ds'].dt.date.astype('datetime64').isin(X_train_holidays)]["ds"],
                           X_test[X_test['ds'].dt.date.astype('datetime64').isin(X_test_holidays)]["ds"]])
holiday_df = pd.DataFrame({"ds": holiday_dates, 
                          "holiday": ['USFederalHoliday'] * len(holiday_dates)}).reset_index().drop("index", axis=1)
holiday_df
````

20. Create an instance of the Prophet model with ```holidays = holiday_df``` the DataFrame we have just created, and train that model on the training set.

````python
model_with_holidays = Prophet(holidays=holiday_df)
model_with_holidays.fit(X_trian)
````

## Predict With Holidays

21. Make predictions for the test set with the holiday model

````python
forecast_holidays = model_with_holidays.predict(X_test)
forecast_holidays.head()
````

## Plot Holiday Effect

22. Plot the different components of the holiday model

````python
fig = model_with_holidays.plot_components(forecast_holidays, figsize=(14, 16))
````

# Error Metrics with Holidays Added

````python
print("The MSE on the test set is : %f" % mean_squared_error(y_true=X_test["y"], y_pred=forecast_holidays["yhat"]))
````

````python
print("The MAE on the test set is : %f" % mean_absolute_error(y_true=X_test["y"], y_pred=forecast_holidays["yhat"]))
````

````python
print("The MAPE on the test set is : %f" % mean_absolute_percentage_error(y_true=X_test["y"], y_pred=forecast_holidays["yhat"]))
````

# Compare Models Just for Holiday Dates
24. Lets plot Forecast model with and without holidays for 4th of July. It should look like the model with holidays is more accurate for this holiday.

````python
july = X_test.loc[(X_test["year"] == 2015) & (X_test["month"] == 7) & (1 <= X_test["dayofmonth"]) & (X_test["dayofmonth"] <= 7)]

models = {'Week of July 4th Forecast vs Actuals non-Holiday Model': (m, forecast), 
          'Week of July 4th Forecast vs Actuals Holiday Model': (model_with_holidays, forecast_holidays)}

f, axes = plt.subplots(2,1,figsize=(14,10), sharex=True)

for idx, element in enumerate(models.items()):
    title, model_info = element
    m, model_forecat = model_info
    july_predict = model_forecat.loc[model_forecat["ds"].isin(july["ds"].values), :]
    axes[idx].scatter(july.ds, july['y'], color='r')
    fig = m.plot(july_predict, ax=axes[idx])
    axes[idx].set_xbound(lower=datetime.strptime('07-01-2015', "%m-%d-%Y"),
                       upper=datetime.strptime('07-07-2015', "%m-%d-%Y"))
    axes[idx].set_title(title)
    fig.show()
````

# Plot Error of Each Forecast
26. Plot the forecast against the target variable for each model on the test set, what do you notice? Are there any observable phenomenon that the model is struggling to predict ?

````python
import seaborn as sns

forecasts = {'Forecast vs Actuals (non-holiday)': forecast, 'Forecast vs Actuals (holiday)': forecast_holidays}

f, axes = plt.subplots(len(forecasts),1,figsize=(14,6*len(forecasts)))

for idx, element in enumerate(forecasts.items()):
    title, forec = element
    data_actual = pd.DataFrame({"ds": X_test["ds"], "y":X_test["y"], "kind": ["Actual"]*len(X_test)})
    data_forecast = pd.DataFrame({"ds": forec["ds"], "y":forec["yhat"], "kind": ["Forecast"]*len(forec)})

    sns.scatterplot(data=pd.concat([data_actual, data_forecast]), x="ds", y="y", hue='kind', s=5, ax=axes[idx])
    axes[idx].set_title(title)

plt.show()
````
